Dummies of market potentials

```python
rmp01_new = rmp01
rmp01_rmp_rv = rmp01_new[["Country Code","Year",'lrmp_rv']]
rmp01_fmp_rv = rmp01_new[["Country Code","Year",'lfmp_rv']]
rmp01_rmp_hm = rmp01_new[["Country Code","Year",'lrmp_hm']]
rmp01_fmp_hm = rmp01_new[["Country Code","Year",'lfmp_hm']]
```

```python
# 1 for high potential (median as the threshold)

rmp01_rmp_rv01 = rmp01_rmp_rv.groupby(by = rmp01_rmp_rv["Country Code"]).mean()
rmp01_rmp_rv01_med = rmp01_rmp_rv01.lrmp_rv.median()
rmp01_rmp_rv01_new = pd.DataFrame({"Country Code": rmp01_rmp_rv01.index.tolist(),
                  "lrmp_rv": rmp01_rmp_rv01["lrmp_rv"].tolist(),
                  })
rmp01_rmp_rv01_new.loc[rmp01_rmp_rv01_new["lrmp_rv"] >= rmp01_rmp_rv01_med, "D_rmp_rv"] = 1
rmp01_rmp_rv01_new.loc[rmp01_rmp_rv01_new["lrmp_rv"] < rmp01_rmp_rv01_med, "D_rmp_rv"] = 0
```

```python
rmp01_fmp_rv01 = rmp01_fmp_rv.groupby(by = rmp01_fmp_rv["Country Code"]).mean()
rmp01_fmp_rv01_med = rmp01_fmp_rv01.lfmp_rv.median()
rmp01_fmp_rv01_new = pd.DataFrame({"Country Code": rmp01_fmp_rv01.index.tolist(),
                  "lfmp_rv": rmp01_fmp_rv01["lfmp_rv"].tolist(),
                  })
rmp01_fmp_rv01_new.loc[rmp01_fmp_rv01_new["lfmp_rv"] >= rmp01_fmp_rv01_med, "D_fmp_rv"] = 1
rmp01_fmp_rv01_new.loc[rmp01_fmp_rv01_new["lfmp_rv"] < rmp01_fmp_rv01_med, "D_fmp_rv"] = 0
```

```python
rmp01_rmp_hm01 = rmp01_rmp_hm.groupby(by = rmp01_rmp_hm["Country Code"]).mean()
rmp01_rmp_hm01_med = rmp01_rmp_hm01.lrmp_hm.median()
rmp01_rmp_hm01_new = pd.DataFrame({"Country Code": rmp01_rmp_hm01.index.tolist(),
                  "lrmp_hm": rmp01_rmp_hm01["lrmp_hm"].tolist(),
                  })
rmp01_rmp_hm01_new.loc[rmp01_rmp_hm01_new["lrmp_hm"] >= rmp01_rmp_hm01_med, "D_rmp_hm"] = 1
rmp01_rmp_hm01_new.loc[rmp01_rmp_hm01_new["lrmp_hm"] < rmp01_rmp_hm01_med, "D_rmp_hm"] = 0
```

```python
rmp01_fmp_hm01 = rmp01_fmp_hm.groupby(by = rmp01_fmp_hm["Country Code"]).mean()
rmp01_fmp_hm01_med = rmp01_fmp_hm01.lfmp_hm.median()
rmp01_fmp_hm01_new = pd.DataFrame({"Country Code": rmp01_fmp_hm01.index.tolist(),
                  "lfmp_hm": rmp01_fmp_hm01["lfmp_hm"].tolist(),
                  })
rmp01_fmp_hm01_new.loc[rmp01_fmp_hm01_new["lfmp_hm"] >= rmp01_fmp_hm01_med, "D_fmp_hm"] = 1
rmp01_fmp_hm01_new.loc[rmp01_fmp_hm01_new["lfmp_hm"] < rmp01_fmp_hm01_med, "D_fmp_hm"] = 0
```

```python
rmp_dummies = pd.merge(rmp01_rmp_rv01_new, rmp01_fmp_rv01_new,
         on=['Country Code'], how='outer')
rmp_dummies = pd.merge(rmp_dummies, rmp01_rmp_hm01_new,
         on=['Country Code'], how='outer')
rmp_dummies = pd.merge(rmp_dummies, rmp01_fmp_hm01_new,
         on=['Country Code'], how='outer')
```

```python
rmp_dummies
```

```python
rmp_dummies = rmp_dummies[["Country Code", "D_rmp_rv", "D_fmp_rv", "D_rmp_hm", "D_fmp_hm"]]
```

Attaching dummies to the dataset

```python
pseries001_merge_new1 = pd.merge(pseries001_merge_new1, rmp_dummies, on=['Country Code'], how='outer')
```
